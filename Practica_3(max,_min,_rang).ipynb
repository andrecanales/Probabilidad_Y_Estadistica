##Histogramas


```python
import random
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50, 100), 2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio2 = round(random.uniform(50, 100), 2)
  fabrica2.append(numero_aleatorio2)
print(fabrica2)
```

```python
import matplotlib.pyplot as plt
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica 1")
```

```python
import matplotlib.pyplot as plt
plt.hist(fabrica2)
plt.title("Produccion diaria de la fabrica 2")
```

```python
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("Valores de la fabrica 1")
print("Media: ", media1)
print("Mediana: ", mediana1)
print("Moda: ", moda1)
```

```python
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print("Valores de la fabrica 2")
print("Media: ", media2)
print("Mediana: ", mediana2)
print("Moda: ", moda2)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows =1, ncols =2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

plt.hist(fabrica1, bins=10, alpha=0.4, color='b', edgecolor = "black")
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label='Mediana')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label='Moda')
plt.xlabel('valor')
plt.ylabel('frecuencia')
plt.title('Produccion diaria de la fabrica 1')
plt.legend()
plt.show()
```

```python
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

plt.hist(fabrica2, bins=20, alpha=0.4, color='b', edgecolor = "black")
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana2, color='r', linestyle='dashed', linewidth=2, label='Mediana')
plt.axvline(moda2, color='y', linestyle='dashed', linewidth=2, label='Moda')
plt.xlabel('valor')
plt.ylabel('frecuencia')
plt.title('Produccion diaria de la fabrica 2')
plt.legend()
plt.show()
```

##Maximo, minimo y rango



```python
maximo = max(fabrica1)
minimo = min(fabrica1)
rango = maximo - minimo
media = mean (fabrica1)
mediana = median(fabrica1)
moda = mode(fabrica1)

print("Estadistica descriptica de la fabrica 1: " "Media =", media,
      "\n Mediana = ", mediana, "\n Moda = ", moda, "\n Maximo =", maximo,
      "\n Minimo = ", minimo, "\n Rango =", rango)
```

```python
maximo2 = max(fabrica2)
minimo2 = min(fabrica2)
rango2 = maximo2 - minimo2
media2 = mean (fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print("Estadistica descriptica de la fabrica 2: " "Media =", media2,
      "\n Mediana = ", mediana2, "\n Moda = ", moda2, "\n Maximo =", maximo2,
      "\n Minimo = ", minimo2, "\n Rango =", rango2)
```
